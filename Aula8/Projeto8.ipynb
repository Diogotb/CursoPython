```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml
```

```python
# Crie gr√°ficos com uma tabela Excel(xlsx)

df = pd.read_excel('Aula8/filmes.xlsx')
 
print(df)
```

```python
# Carregar dados do arquivo YAML
with open('Aula8/dados.yaml', 'r') as file:
    dados = yaml.safe_load(file)

# Criar DataFrame com os dados
df = pd.DataFrame(dados)

# Plotar histograma da idade com Matplotlib
plt.figure(figsize=(8, 6))  # Criar uma figura com tamanho espec√≠fico
plt.hist(df['idade'], bins=5, color='skyblue', edgecolor='black')  # Plotar histograma
plt.title('Histograma da Idade - Matplotlib')  # Definir t√≠tulo do gr√°fico
plt.xlabel('Idade')  # Definir r√≥tulo do eixo x
plt.ylabel('Frequ√™ncia')  # Definir r√≥tulo do eixo y
plt.grid(True)  # Habilitar grade no gr√°fico
plt.show()  # Mostrar o gr√°fico





```

```python
# Plotar gr√°fico de dispers√£o da altura e idade com Seaborn
plt.figure(figsize=(8, 6))  # Criar uma figura com tamanho espec√≠fico
sns.scatterplot(data=df, x='altura', y='idade', color='purple', alpha=0.5)  # Plotar gr√°fico de dispers√£o com Seaborn
plt.title('Gr√°fico de Dispers√£o da Altura e Idade - Seaborn')  # Definir t√≠tulo do gr√°fico
plt.xlabel('Altura')  # Definir r√≥tulo do eixo x
plt.ylabel('Idade')  # Definir r√≥tulo do eixo y
plt.grid(True)  # Habilitar grade no gr√°fico
plt.show()  # Mostrar o gr√°fico
```

```python
# Plotar gr√°fico de barras das notas com Seaborn
plt.figure(figsize=(8, 6))  # Criar uma figura com tamanho espec√≠fico
sns.barplot(data=df, x='alunos', y='notas', palette='pastel')  # Plotar gr√°fico de barras com Seaborn
plt.title('Gr√°fico de Barras das Notas - Seaborn')  # Definir t√≠tulo do gr√°fico
plt.xlabel('Alunos')  # Definir r√≥tulo do eixo x
plt.ylabel('Notas')  # Definir r√≥tulo do eixo y
plt.grid(True)  # Habilitar grade no gr√°fico
plt.show()  # Mostrar o gr√°fico
```

```python
# Plotar gr√°fico de dispers√£o da altura e idade com Matplotlib
plt.figure(figsize=(8, 6))  # Criar uma figura com tamanho espec√≠fico
plt.scatter(df['altura'], df['idade'], color='green', alpha=0.5)  # Plotar gr√°fico de dispers√£o
plt.title('Gr√°fico de Dispers√£o da Altura e Idade - Matplotlib')  # Definir t√≠tulo do gr√°fico
plt.xlabel('Altura')  # Definir r√≥tulo do eixo x
plt.ylabel('Idade')  # Definir r√≥tulo do eixo y
plt.grid(True)  # Habilitar grade no gr√°fico
plt.show()  # Mostrar o gr√°fico
```
